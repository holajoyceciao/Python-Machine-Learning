```python
import pandas as pd

from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
# the features of data
X = iris.data
# the data for the prediction
y = iris.target 
# print('X:', X)
# print(y)

feature_names = iris.feature_names
target_names = iris.target_names
# print('X column names:', feature_names)
print('y column name:', target_names)

data = pd.DataFrame(X, columns=[feature_names])
data['target'] = y
data
```

### Split the data

```python
data
```

```python
X = data.drop('target', axis=1)
y = data['target']
```

```python
from sklearn.model_selection import train_test_split

# Spliting to training data & testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X.shape)
print(X_train.shape)
print(X_test.shape)
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

### Create A Model

```python
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier

# using the algorithm
knn = KNeighborsClassifier(n_neighbors=3)

# fit(): train the machines with given training data
knn.fit(X_train, y_train)

# predict(): using the true data to predict
y_pred = knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
from sklearn import metrics

# The accuracy of the model's output
print(metrics.accuracy_score(y_test, y_pred)) 
```

```python
sample=[[3,5,4,2],[2,3,5,4]]
predictions = knn.predict(sample)
print(predictions)
pred_species = [iris.target_names[p] for p in predictions]
print('predictions: ',pred_species)
```

```python
# save the model using joblib
import joblib 

joblib.dump(knn, 'mlbrain.joblib')
```

```python
# use model 
model = joblib.load('mlbrain.joblib')
model.predict(X_test)
sample=[[3,5,4,2],[2,3,5,4]]
predictions = model.predict(sample)
pred_species = [iris.target_names[p] for p in predictions]
print('predictions: ',pred_species)
```
